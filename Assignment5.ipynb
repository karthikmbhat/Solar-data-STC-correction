```python
Isc = 7.987649
beta = (-0.32*37.08)/100
T2 = 273.15+25
T1 = 273.15+56.5
G2 = 1000
G1 = 919.877
for i in range(0, 111):
    V1=df.iloc[:i,0]
    V2=V1+beta*(T2-T1)
print V1
print V2
```
